## 测试regex正则表达式

```python
import regex
pat = regex.compile(r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}++| ?\p{N}++| ?[^\s\p{L}\p{N}]++|\s++$|\s+(?!\S)|\s""")
print(pat.findall('Hello, world!'))
print(pat.findall('the cat ate'))
print(pat.findall("I'm running"))
```
